```python
file1    = 'data/mcec_train.csv'
df_train = pd.read_csv(file1)

file2    = 'data/mcec_dev_translated.pkl'
df_dev   = pd.read_pickle(file2)

print(df_train.shape, df_dev.shape)
```

```python
df_train = df_train.rename( columns={'Text': 'text', 'Emotion': 'emotion',} )
```

```python
source     = df_dev['text'].tolist()
res        = dict()
keep_going = True
translator = Translator()
max_iter   = 0

while keep_going:
        
    # repeat only 10 times
    max_iter += 1
    print(f'\nWHILE LOOP #{max_iter}: translated {len(res)} sentences\n')
    if max_iter > 10:
        print('Max number of iterations exceeded. Stopping')
        break
        
    # iterate once, translate what can be translated
    for sent in source:
        if sent in res:
            continue
        try:
            translated = translator.translate(sent, src='ur', dest='en')
            res[sent]  = translated.text
        except:
            print(f'Exception (inside for loop). Translated a total of {len(res)} sentences')
        
    # if not everything is translated, there will be another while loop
    if len(res) == len(source):
        keep_going = False
        break
    print(f'Outside for loop: translated {len(res)} sentences')
```

```python
df_dev['translated_ur'] = df_dev['text'].map( res )
df_dev.isna().sum()
```
