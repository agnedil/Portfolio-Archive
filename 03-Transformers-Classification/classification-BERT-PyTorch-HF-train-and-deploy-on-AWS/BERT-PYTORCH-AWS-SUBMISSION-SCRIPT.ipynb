```python
import os
import sagemaker
from sagemaker.pytorch.estimator import PyTorch
import time
```

```python
os.environ["AWS_REGION"] = ""
os.environ["AWS_DEFAULT_REGION"] = ""
os.environ["AWS_ACCESS_KEY_ID"] = ""
os.environ["AWS_SECRET_ACCESS_KEY"] = ""
os.environ["AWS_SESSION_TOKEN"] = ""
```

```python
OUTPUT_PATH   = 's3://'
CODE_LOCATION = 's3://'
IAM_ROLE      = ''
SOURCE_DIR    = '/Users/'
TRAIN_FILE    = 'train_binary_pytorch.py'


estimator = PyTorch(
                        entry_point       = TRAIN_FILE,                 
                        source_dir        = SOURCE_DIR,                        # this helps find requirements.txt
                        role              = IAM_ROLE,
                        instance_count    = 1,
                        instance_type     = 'ml.g4dn.8xlarge',
                        framework_version = '1.9.0',
                        py_version        = 'py38',
                        script_mode       = True,
                        hyperparameters   = {},
                        output_path       = OUTPUT_PATH,
                        model_dir         = OUTPUT_PATH,
                        #code_location     = CODE_LOCATION,
)
```

```python
time_stamp = time.strftime("%Y%m%dT%H%M")
estimator.fit( job_name = f"binary-bert-torch-{time_stamp}" )
```
